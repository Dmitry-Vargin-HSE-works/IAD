## Работа с данными. Pandas <a id="pandas"></a>

**Pandas** - библотека для анализа данных, которая предоставляет широкие возможности для работы с табличными данными и временными рядами. Основана на numpy.  
Для работы с данными в библиотеке используются две основные структуры данных - **Series** (для одномерных данных одного типа) и **Data Frame** (для табличных данных вида *наблюдения* $\times$ *признаки*).  
  
У Pandas отличная [документация](https://pandas.pydata.org/docs/user_guide/index.html), в которой описывается широкий функционал библиотеки. Обязательно к прочтению [краткое введение](https://pandas.pydata.org/docs/user_guide/10min.html#min).  

### Задача прогнозирования оттока клиентов

Классическая задача Data Science. Заключается в следующем: 
 - Выявить основные причины, почему уходят пользователи/клиенты (анализ данных)
 - Построить модель, которая своевременно сингнализировала бы о том, что конкретный пользователь/клиент скоро уйдет с высокой долей вероятности (машинное обучение)
 
В качестве примера будем использовать данные по оттоку клиентов одной американской компании телеком-оператора. Оригинал данных можно посмотреть [здесь](https://bigml.com/user/francisco/gallery/dataset/5163ad540c0b5e5b22000383).  
Даные представляют клиентов компании, которые характеризуются конечныс набором признаков (описаны ниже). Для каждого клиента известно, ушел он или нет (признак `Churn`).

### Data Frame <a id="df"></a>

Структура данных Data Frame представляет из себя "матрицу", в которой все элементы в одном столбце принадлежать одному типу данных. Типы данных самих столбцов могут различаться.  
Такая структура отлично подходит (и была создана) для представления данных типа *наблюдения* $\times$ *признаки*: каждая строчка в data frame соотвествует наблюдению (объекту), а столбцы содержат значения признаков, которыми эти объекты характеризуются.

У каждой строки и каждого столбца в data frame есть номер - как в обычной матрице. По номерам можно выполнять индексацию и сабсеты, как в матрицах.  
Кроме того, у каждой строки и кажого столбца есть индексы - "имена" -, по которым к ним также можно обращаться.

Создадим data frame из матрицы:

````python
np.random.seed(12345)
M = np.random.randint(low=10, high=99, size=(5,3))
M
````

````python
df = pd.DataFrame(M, columns=['one', 'two', 'three'])
df
````

### Subsetting dataframe <a id="subset"></a>

К элементам data frame можно обращаться по индексам (именам) строк и столбцов, по их номерам, а также через булевую маску строк. 

**Сабсет по одному или нескольким признакам:**

````python
df['Total day minutes'].mean() # среднее значение по столбцу
````

````python
df[['Total day minutes', 'Total eve minutes', 'Total night minutes']].max(axis=0) # наибольшее по каждому столбцу
````

<br>  
**Сабсет по индексам и признакам:**

````python
df.loc[[1, 176, 2345]]
````

````python
df.loc[[1, 176, 2345], ['State', 'Churn']]
````

<br>  
**Сабсет по номерам:**

````python
df.iloc[[1, 176, 2345], [0, 19]]
````

````python
df.iloc[[1, 176, 2345]]
````

````python
df.iloc[:, [0, 19]].nunique()
````

<br>  
**Срезы:**

````python
df.loc[0:5, 'Total day minutes':'Total day charge'] # последний элемент включается
````

````python
df.iloc[0:5, 6:9] # последний элемент НЕ включается
````

<br>  
**Булевая маска aka сабсет по условию:**

````python
cond = df['Churn'] == True
cond.value_counts()
````

````python
print(df[cond]['Total day minutes'].mean())
````

````python
print(df[df['Churn'] == False]['Total day minutes'].mean())
````

При сабсете по условию можно также использовать логические операторы для сложных условий. Но...  
Во-первых, необходимо использовать побитовый `&` и `|` вместо привычных `and` и `or`. Во-вторых, условия обязательно должны быть заключены в скобки.  
Подробнее об этом можно почитать [здесь](https://stackoverflow.com/a/21415990).

````python
print(df[(df['Churn'] == True) & (df['International plan'] == 'Yes')]['Total day minutes'].mean())
print(df[(df['International plan'] == 'Yes') | ((df['Voice mail plan'] == 'Yes'))]['Total day minutes'].mean())
````

### Изменение data frame <a id="df_mod"></a>

Добавление столбца:

````python
total_calls = df['Total day calls'] + df['Total eve calls'] \
    + df['Total night calls'] + df['Total intl calls']
````

````python
df['New col'] = total_calls
````

Добавлени столбца на конкретную позицию:

````python
df.insert(loc=2, column='Another new col', value=total_calls) 
````

````python
df.head()
````

<br>  
Изменение значений - просто через присваивание:

````python
df.loc[0:3, 'New col'] = [5,6,7,8]
````

````python
df[df['Area code'] == 415]['Another new col'] = 3 # потому что копия, а не ссылка
````

````python
id(df[df['Area code'] == 415]['State'])
````

````python
id(df.loc[df['Area code'] == 415, 'State'])
````

````python
df.loc[df['Area code'] == 415, 'Another new col'] = 3 # now it's ok
````

````python
df.head()
````

<br>  
Удаление столбца:

````python
df.drop(['New col', 'Another new col'], axis=1, inplace=True) # axis=0 will do for rows
````

**NB:** При удалении строк номера строк, которые были ниже удаленных, изменятся (сдвинутся выше), но индексы (имена) строк **не изменятся** - это может привести к коллизиям при выполнении некоторых операций. Чтобы привести индексы строк в соотвествие с номерами, можно использовать метод `reset_index()`.

````python
df.head()
````

<br>  
Также есть ряд методов, позволяющих изменять данные inplace. Например, `replace()`:

````python
df['State'].replace('KS', 'ks', inplace=True)
df.head()
````

````python
df.replace({'Voice mail plan': {'No' : False, 'Yes' : True}}, inplace=True)
df.head()
````

Аналогично, но через присваивание и `map()`:

````python
df['International plan'] = df['International plan'].map({'No' : False, 'Yes' : True})
df.head()
````

### Начальная модель (baseline)

**Задание:**  
* Объединим наши два инсайта в один признак: подключил ли пользователь роуминг и звонил ли в тех поддержку более 3-х раз.  
Теперь можно попробовать использовать этот признак для классификации: пусть если пользователь и подключил роуминг, и звонил в тех поддержку, то будем считать, что он уйдет. Иначе считаем, что пользователь лоялен.  
* Оцените качество такого решения: как много клиентов мы определим в правльный класс - ушел или остался. 

````python
pd.crosstab(df['Many_service_calls'] & df['International plan'] , 
            df['Churn'], normalize=True, margins=True)
````

При таком подходе мы получим точность 0.8523 + 0.0057 = 0.858. Что лишь немного больше наивной модели (если мы всех пользователей будем определять лояльными - 0.8551). Но это лишь первый шаг, при котором мы уже смогли правильно классифицировать 0.0057/0.14 = 0.04 уходящих клиентов. При этом мы всего лишь немного посмотрели на наши данные.
